```python
city_count= data['City'].value_counts()
city_count = city_count.reset_index()
city_count
```

```python
data2 = data.groupby('City').agg({'Quantity': 'sum', 'Total': 'sum', 'gross income': 'sum'})
data2.columns = [ 'sum_of_Quantity', 'sum_of_Total', 'sum_of_profit']
data2 = data2.reset_index()
data2
```

```python
data3 = pd.merge(city_count, data2, on='City')
data3 = data3.rename(columns={'count':'sum_of_purchase'})
data3
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming data3 is already created and contains the data
# If not, recreate it using the previous code

# Calculate percentages
total_purchases = data3['sum_of_purchase'].sum()
data3['percentage'] = (data3['sum_of_purchase'] / total_purchases) * 100

# Create the pie chart
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
plt.pie(data3['percentage'], labels=data3['City'], autopct='%1.1f%%', startangle=90)
plt.title('Tỷ lệ phần trăm tổng số lượt mua theo thành phố')
plt.show()
```

# Biểu đồ doanh số theo tháng

```python
data['CustomerID_temp'] = data['Invoice ID'].astype(str) + '_' + data['City'] + '_' + data['Customer type'] + '_' + data['Gender'] + '_' + data['Date'].astype(str)
data['Date'] = pd.to_datetime(data['Date'])
data
```

```python
import matplotlib.pyplot as plt

# Tạo các cột mới cho 'Month', 'Quarter', and 'Year'
data['Month'] = data['Date'].dt.month
data['Year'] = data['Date'].dt.year


# Phân tích doanh số theo ngày
daily_sales = data.groupby('Date')['Total'].sum().reset_index()
plt.figure(figsize=(12, 6))
plt.plot(daily_sales['Date'], daily_sales['Total'])
plt.title('Doanh số bán hàng theo ngày')
plt.xlabel('Ngày')
plt.ylabel('Doanh số')
plt.show()

# Phân tích doanh số theo tháng
monthly_sales = data.groupby(['Year', 'Month'])['Total'].sum().reset_index()
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['Month'], monthly_sales['Total'])
plt.title('Doanh số bán hàng theo tháng')
plt.xlabel('Tháng')
plt.ylabel('Doanh số')
plt.show()


```

+ Doanh số giảm mạnh từ tháng 1 đến tháng 2.
+ Đến tháng 2, doanh số đạt mức thấp nhất (dưới 98.000).
+ Tăng trở lại vào tháng 3, nhưng chưa bằng mức doanh số tháng 1.
+ Nguyên nhân doanh số giảm mạnh tháng 2 có thể do:  - Sau kỳ nghỉ lễ hoặc Tết, làm giảm nhu cầu mua sắm.
                                                     - Thị trường chậm lại theo chu kỳ kinh doanh.
                                                     - Ảnh hưởng từ yếu tố bên ngoài như kinh tế, thời tiết, chính sách bán hàng.
   Doanh số phục hồi vào tháng 3 có thể do:
                                            - Các chương trình khuyến mãi, kích cầu sau kỳ nghỉ.
                                            - Tăng trưởng tự nhiên theo xu hướng kinh doanh.

```python
data.set_index('Date', inplace=True)
```

```python
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
data = data.reset_index()  
train_data = data[:-30]
test_data = data[-30:]

# Xây dựng mô hình ARIMA (ví dụ: ARIMA(5,1,0))
model = ARIMA(train_data['Total'], order=(5,1,0))
model_fit = model.fit()

# Dự đoán trên tập kiểm tra
predictions = model_fit.predict(start=len(train_data), end=len(data)-1)

# Đánh giá mô hình
rmse = np.sqrt(mean_squared_error(test_data['Total'], predictions))
print('RMSE:', rmse)

# Vẽ biểu đồ dự đoán và giá trị thực tế
plt.plot(test_data.index, test_data['Total'], label='Thực tế')
plt.plot(test_data.index, predictions, label='Dự đoán')
plt.legend()
plt.show()
```
